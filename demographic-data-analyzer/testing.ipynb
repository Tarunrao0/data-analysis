#### percentage of people who have a bachelor's degree

# Problem 4

#### people with advanced education making more than 50k

```python
advanced_education = ['Bachelors', 'Masters', 'Doctrate']
advanced_edu_df = df[df['education'].isin(advanced_education)]
high_earners_df = advanced_edu_df[advanced_edu_df['salary'] == '>50K']
percentage = (high_earners_df.shape[0] / advanced_edu_df.shape[0]) * 100
percentage
```

```python
df.groupby(['education']).size()
```

```python
lower_edu_df = df[~df['education'].isin(advanced_education)]
high_earners_df_low = lower_edu_df[lower_edu_df['salary'] == '>50K']
percentage = (high_earners_df_low.shape[0] / lower_edu_df.shape[0]) * 100
percentage
```

# Problem 6

#### number of people who work minimum hours a week
